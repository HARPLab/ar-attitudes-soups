```python
# function to return ordered list from dict
def ordlist(dict):
  returnlst = np.zeros(5)
  for x in dict:
    if x == "Extremely uncomfortable":
      returnlst[4] = dict[x]
    if x == "Somewhat uncomfortable":
      returnlst[3] = dict[x]
    if x == "Neither comfortable nor uncomfortable":
      returnlst[2] = dict[x]
    if x == "Somewhat comfortable":
      returnlst[1] = dict[x]
    if x == "Extremely comfortable":
      returnlst[0] = dict[x]
  return list(returnlst)
```

```python
# Audio you intentionally provide
s = data['Q59_1'].value_counts(sort = False)
w = data['Q59_2'].value_counts(sort = False)
p = data['Q59_3'].value_counts(sort = False)
sv = ordlist(dict(s))
wv = ordlist(dict(w))
pv = ordlist(dict(p))
index = ['Extremely comfortable', 'Somewhat comfortable', 'Neither comfortable nor uncomfortable', 'Somewhat uncomfortable', 'Extremely uncomfortable']
# bar plot
df = pd.DataFrame({'Snapchat': sv,
                   'PostureScreen': pv,
                   'WarbyP': wv}, index=index)
ax = df.plot.bar(stacked = True)
```
